# Ejercicio: Visualizador de ventas en un grafico

**1 -> Importar la libreria de Matplotlib**

```python
import matplotlib.pyplot as plt
```

**2 -> Se generan las listas con los datos ejemplos**

```python
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
ventas = [1500, 2500, 3000, 4000, 3500, 5000]
```

**3 -> Crear la gr√°fica**

```python
# Se crea una imagen, con las medidas de 10 pulgadas de ancho y 6 pulgadas de altura.
plt.figure(figsize=(10, 6))
# Se crea un grafico de lineas, se a√±aden los valores de meses en el eje x, se a√±aden los valor de ventas en el eje y
# marker, define el marcador que usara en el punto del grafico
# linestyle, indica el estilo de la linea que conecta los puntos, linea continua
# color, indica el color de la linea, en este caso b es azul
plt.plot(meses, ventas, marker='o', linestyle='-', color='b')

# Se define el titulo de la grafica, fontsize, indica el tama√±o del texto
plt.title('Crecimiento de Ventas en el Primer Semestre', fontsize=16)
# Se a√±ade una descripcion de la columna x
plt.xlabel('Meses', fontsize=14)
# Se a√±ade una descripcion de la columna y
plt.ylabel('Ventas ($)', fontsize=14)

# A√±ade una oja cuadriculada al fondo
plt.grid(True)
# Mostrar grafica
plt.show()
```
