# Ejercicio: Genera una grafica donde se muestren los titulos de los equipos en Colombia

### 1 -> Se crea un diccionario el cual tenga el nombre de los equipos de Colombia y su cantidad de titulos.

```python
equiposColombianos = {
    "Atlético Bucaramanga": 1,
    "América de Cali": 15,
    "Millonarios": 16,
    "Independiente Medellín": 6,
    "Unión Magdalena": 1,
    "Junior": 10,
    "Santa Fe": 9,
    "Deportes Quindío": 1,
    "Once Caldas": 4,
    "Deportivo Cali": 10,
    "Tolima": 3,
    "Cúcuta Deportivo": 1,
    "Atlético Nacional": 17,
    "Boyacá Chicó": 1,
    "Deportes Tolima": 3,
    "La Equidad": 1
}
```

### 2 -> Se procede a generar una lista donde se saquen el nombre de los equipos y los titulos

```python
equipos = list(equiposColombianos.keys())
titulos = list(equiposColombianos.values())
```

## Integracion de Matplotlib

### 3 -> Importacion libreria Matplotlib

```python
import matplotlib.pyplot as plt
```

### 4 -> Se crea el grafico.

```python
# Se genera una figura, en la cual se define su tamaño, ancho de 10 y altura de 8 pulgadas
plt.figure(figsize=(10, 8))
# Se define el grafico a utilizar (horizontal), se define los nombre de los equipos de los ejes junto a sus titulos y el color de la barra
plt.barh(equipos, titulos, color='skyblue')
# Se añade una etiqueta en el eje x, donde se informa la cantidad de titulos
plt.xlabel('Número de Títulos')
# Se añade una etiqueta en el eje y, donde se informa el nombre de los equipos
plt.ylabel('Equipos')
# Se le asigna un titulo al grafico
plt.title('Títulos de Liga de Equipos Colombianos')
# Mostrar grafico
plt.show()
```
