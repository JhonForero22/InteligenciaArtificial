* Evolución del Consumo:

Gráfico de línea del consumo de cerveza y pollo asado a lo largo de los años: Muestra cómo ha cambiado el consumo de cerveza y pollo asado con el tiempo.



```python
plt.figure()
plt.plot(df_Oktoberfest['year'], df_Oktoberfest['beer_consumption'], label='Consumo de Cerveza (L)', marker='o') #Grafico con la funcion con el valor de x y y
plt.plot(df_Oktoberfest['year'], df_Oktoberfest['roast_chicken_consumption'], label='Consumo de Pollo Asado (kg)', marker='o') #Grafico con la funcion con el valor de x y y
plt.title('Consumo de Cerveza y Pollo Asado a lo largo de los años') #Titulo del grafico
plt.xlabel('Año') # Etiqueta para el eje X
plt.ylabel('Consumo') # Etiqueta para el eje Y
plt.legend()
plt.grid(True)
plt.show()
```

* Análisis del Precio vs. Consumo:

Gráfico de dispersión del precio de la cerveza frente al consumo de cerveza: Analiza si hay alguna correlación entre el precio de la cerveza y la cantidad consumida.



```python
plt.figure()
plt.scatter(df_Oktoberfest['beer_price'], df_Oktoberfest['beer_consumption'], color='blue', marker='o')
plt.title('Relación entre el Precio de la Cerveza y su Consumo')
plt.xlabel('Precio de la Cerveza (USD/L)')
plt.ylabel('Consumo de Cerveza (L/persona/año)')
# Mostrar el gráfico
plt.grid(True)
plt.show()
```

* Comparación de Precios y Consumo:

Gráfico de barras comparativo del precio promedio de la cerveza y del pollo asado a lo largo de los años: Ve cómo han cambiado los precios de estos productos con el tiempo.

```python
plt.figure()
plt.plot(df_Oktoberfest['year'], df_Oktoberfest['beer_price'], color='green', label='Precio Cerveza', marker='o')
plt.plot(df_Oktoberfest['year'], df_Oktoberfest['roast_chicken_price'], color='red',label='Precio Pollo', marker='o')
plt.title('Comparativo del precio promedio de la cerveza y del pollo asado al largo de los años')
plt.xlabel('Año')
plt.ylabel('Promedio cerveza/polloAsado')
# Mostrar el gráfico
plt.grid(True)
plt.legend()
plt.show()
```
