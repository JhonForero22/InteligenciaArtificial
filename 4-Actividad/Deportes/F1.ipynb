# Conteo por Nacionalidad
Puedes crear un gráfico de barras que muestre la cantidad de constructores por nacionalidad. Esto te permitirá ver qué nacionalidades están más representadas en tu dataset.

```python
# Conteo de constructores por nacionalidad
nacionalidad_counts = df_f1['nationality'].value_counts()

plt.figure(figsize=(10, 6))
nacionalidad_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Nacionalidad')
plt.ylabel('Cantidad de Constructores')
plt.title('Cantidad de Constructores por Nacionalidad')
plt.xticks(rotation=45)
plt.show()
```
