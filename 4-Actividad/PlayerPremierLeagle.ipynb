* Seleccionar el archivo y asignarle un nombre

```python
df = pd.read_csv('premier-player-23-24.csv')
```

* Imprimir los datos del archivo csv

```python
print(df.head())  # Muestra las primeras 5 filas del archivo CSV
```

* Crear una funcion

```python
df_filtrado = df[df['Gls'] > 10] #Jugadores con mas de 10 goles
```

* Filtrar por jugadores que uno quiera buscar

```python
nombres_deseados = ['Luis Díaz', 'Daniel Muñoz']
df_filtrado = df[df['Player'].isin(nombres_deseados)]
```

* Generar el grafico

```python
plt.figure(figsize=(10, 6)) #Medidas de la figura
plt.plot(df_filtrado['Player'], df_filtrado['Gls'], label='Datos X vs Y') #Grafico con la funcion con el valor de x y y
plt.xlabel('Jugador') # Etiqueta para el eje X
plt.ylabel('Goles') # Etiqueta para el eje Y
plt.title('Gráfico de Líneas, Goles marcados') #Titulo del grafico
plt.xticks(rotation=90) #Rotacion titulo eje x
plt.legend()
plt.grid(True)
plt.show()
```

* Contar cuentos jugadores hay de un pais

```python
# Inicializar contador
cantidadJugadores = 0
pais_deseado = 'co COL'

# Recorrer el DataFrame usando itertuples y filtrar por el país deseado
for row in df.itertuples():
    if row.Nation == pais_deseado:
        cantidadJugadores += 1

print(f'Cantidad de jugadores del país deseado ({pais_deseado}): {cantidadJugadores}')
```

* Mostrar informacion de los jugadores de un pais

```python
# Inicializar contador
pais_deseado = 'co COL'

# Recorrer el DataFrame usando itertuples y filtrar por el país deseado
for row in df.itertuples():
    if row.Nation == pais_deseado:
        print("---")
        print(f"Player: {row.Player}")
        print(f"Age: {row.Age}")
        print(f"Goals: {row.Gls}")
        print(f"Team: {row.Team}")
        print(f"Country: {row.Nation}")
        print("---")
```
