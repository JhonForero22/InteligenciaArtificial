# Gráfico de líneas (Line Plot)
Evolución del rendimiento según la temperatura o la humedad: Puedes hacer un gráfico de líneas para mostrar cómo cambia el rendimiento (label) con respecto a la temperatura o la humedad a lo largo del tiempo o de diferentes muestras.
Eje X: Temperatura o Humedad.
Eje Y: label (rendimiento).

```python
# Configura el tamaño de la figura
plt.figure(figsize=(10, 6))

# Dibuja la línea para la relación entre temperatura y rendimiento
plt.plot(dfAgricultura['temperature'], dfAgricultura['label'], marker='o', linestyle='-', color='b', label='Temperatura vs. Rendimiento')

# Opcional: Añadir la línea para la relación entre humedad y rendimiento
plt.plot(dfAgricultura['humidity'], dfAgricultura['label'], marker='x', linestyle='--', color='g', label='Humedad vs. Rendimiento')

# Añadir títulos y etiquetas
plt.title('Evolución del Rendimiento según Temperatura y Humedad')
plt.xlabel('Temperatura / Humedad')
plt.ylabel('Rendimiento (Label)')
plt.legend()

# Mostrar el gráfico
plt.show()
```

# Gráfico de barras (Bar Plot)
Comparación de rendimiento por categorías: Si el label es categórico (por ejemplo, tipos de cultivos o zonas), puedes hacer un gráfico de barras para comparar el rendimiento entre diferentes categorías.
Eje X: Categorías (por ejemplo, tipo de cultivo).
Eje Y: label (rendimiento).

Paso 3: Asegurarse de que las columnas sean del tipo correcto

```python
# Asegúrate de que la columna de categorías sea de tipo categórico
dfAgricultura['label'] = dfAgricultura['label'].astype('category')

# Asegúrate de que la columna de rendimiento sea numérica
dfAgricultura['rainfall'] = pd.to_numeric(dfAgricultura['rainfall'], errors='coerce')
```

Paso 4: Manejar valores faltantes

```python
# Elimina filas con valores faltantes en las columnas relevantes
df = dfAgricultura.dropna(subset=['label', 'rainfall'])
```

Paso 5: Crear el gráfico de barras

```python
plt.figure(figsize=(10, 6))

# Crea el gráfico de barras
sns.barplot(x='label', y='rainfall', data=df, palette='viridis')

# Añadir títulos y etiquetas
plt.title('Comparación de Rendimiento por Categorías')
plt.xlabel('Categorías (Label)')
plt.ylabel('Rendimiento')
plt.xticks(rotation=90)

# Mostrar el gráfico
plt.show()
```
